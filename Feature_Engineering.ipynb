# Missing values

```python
## Let us capture all the nan values
## First lets handle Categorical features which are missing
features_nan=[feature for feature in df_hp.columns if df_hp[feature].isnull().sum()>1 and df_hp[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(df_hp[feature].isnull().mean(),4)))
```

```python
## Replace missing value with a new label
def replace_cat_feature(df_hp,features_nan):
    data=df_hp.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

df_hp=replace_cat_feature(df_hp,features_nan)

df_hp[features_nan].isnull().sum()
```

```python
df_hp.head()
```

# Numerical variables

```python
## Now lets check for numerical variables the contains missing values
numerical_with_nan=[feature for feature in df_hp.columns if df_hp[feature].isnull().sum()>1 and df_hp[feature].dtypes!='O']

## We will print the numerical nan variables and percentage of missing values

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(df_hp[feature].isnull().mean(),4)))
```

```python
## Replacing the numerical Missing Values with mean

for feature in numerical_with_nan:
    ## We will replace by using median since there are outliers
    median_value=df_hp[feature].median()
    
    ## create a new feature to capture nan values
    df_hp[feature+'nan']=np.where(df_hp[feature].isnull(),1,0)
    df_hp[feature].fillna(median_value,inplace=True)
    
df_hp[numerical_with_nan].isnull().sum()
```

# Handling Numerical variables 
taking log as data of numerical variables are skewed

```python
import numpy as np
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    df_hp[feature]=np.log(df_hp[feature])
```

# Temporal VAriables

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    df_hp[feature]=df_hp['YrSold']-df_hp[feature]
```

# Handling categorical variables
We will remove categorical variables that are present less than 1% of the observations


```python
categorical_features=[feature for feature in df_hp.columns if df_hp[feature].dtype=='O']
```

```python
for feature in categorical_features:
    temp=df_hp.groupby(feature)['SalePrice'].count()/len(df_hp)
    #print(temp)
    temp_df=temp[temp>0.01].index
    #print(temp_df)
    df_hp[feature]=np.where(df_hp[feature].isin(temp_df),df_hp[feature],'Rare_var')
    print(df_hp[feature])
```

```python
for feature in categorical_features:
    labels_ordered=df_hp.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df_hp[feature]=df_hp[feature].map(labels_ordered)
   
```

```python
feature_scale=[feature for feature in df_hp.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df_hp[feature_scale])
```
